```python
# remove last 3 characters
df['Date'] = df['Date'].map(lambda x: str(x)[:-3])
```

```python
# convert rus date (str) into eng datetype
def rus_to_eng(x):

    months_rus = ['января', 'февраля', 'марта', 'апреля', 'мая', 'июня', 
                  'июля', 'августа', 'сентября', 'октября', 'ноября', 'декабря']

    months_eng = ['January', 'February', 'March', 'April', 'May', 'June',
                  'July', 'August', 'September', 'October', 'November', 'December']
    
    for idx, month_rus, month_eng in zip(range(len(months_rus)), months_rus, months_eng):
        if month_rus in x:
            lst = x.split()
            lst[1] = months_eng[idx]
            datum = '-'.join(lst)
            datum = datetime.strptime(datum, "%d-%B-%Y")
            return datum
```

```python
# apply function
df['Date'] = df['Date'].apply(rus_to_eng)
df.head()
```
